```python
# 필요한 패키지 import 하기
import os # 환경 변수나 디렉터리, 파일 등의 OS 자원을 제어할 수 있게 해주는 모듈
import cv2 # OpenCV라이브러리 → 컴퓨터 비전 관련 프로그래밍을 쉽게 할 수 있도록 도와주는 라이브러리
import matplotlib.pyplot as plt # 다양한 데이터를 많은 방법으로 도식화 할 수 있도록 하는 라이브러리
import numpy as np # 다차원 배열을 쉽게 처리하고 효율적으로 사용할 수 있도록 하는 라이브러리
import dlib # 이미지 처리 및 기계 학습, 얼굴인식 등을 할 수 있는 c++ 로 개발된 고성능의 라이브러리
from PIL import Image

def add_cat_whiskers_dlib(human_img_path, whisker_img_path, output_path='output.png'):
    # 이미지 불러오기
    img_bgr = cv2.imread(human_img_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(img_rgb).convert("RGBA")

    # 얼굴 검출기 및 랜드마크 예측기 초기화
    predictor_path = 'shape_predictor_68_face_landmarks.dat'
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(predictor_path)

    # 얼굴 검출
    dets = detector(img_rgb, 1)
    if len(dets) == 0:
        print("❌ 얼굴을 찾을 수 없습니다.")
        return

    # 수염 이미지 불러오기
    whisker_img = Image.open(whisker_img_path).convert("RGBA")

    for d in dets:
        # 랜드마크 추출
        landmarks = predictor(img_rgb, d)
        points = [(p.x, p.y) for p in landmarks.parts()]

        # 코 아래 중심 좌표 = landmark 33번
        nose_x, nose_y = points[30]

        # 수염 이미지 크기 조정 (얼굴 폭 비례)
        face_width = d.right() - d.left()
        new_width = int(face_width * 0.6)
        new_height = int(new_width * whisker_img.height / whisker_img.width)
        whisker_resized = whisker_img.resize((new_width, new_height))

        # 수염 이미지 붙이기 (중심 정렬)
        wx = nose_x - new_width // 2
        wy = nose_y - new_height * 2 // 5  # 코보다 약간 위쪽에 오도록 조정

        img_pil.paste(whisker_resized, (wx, wy), whisker_resized)

    img_pil.save(output_path)
    print(f"✅ 결과 이미지 저장됨: {output_path}")
    plt.imshow(img_pil)
    plt.axis('off')  # 축 제거 (선택)
    plt.show()
```

```python
add_cat_whiskers_dlib(
    human_img_path='images/image.png',
    whisker_img_path='images/cat-whiskers.png',
    output_path='images/with_whiskers.png'
)
```

```python
# 수염 각도가 얼굴 각도와 맞지 않는 문제 발생
# 사각형과 랜드마크를 그려서 분석 진행
```

```python
def show_face_rectangles(image_path):
    # 1. 이미지 로드
    img_bgr = cv2.imread(image_path)
    if img_bgr is None:
        print("❌ 이미지 파일을 찾을 수 없습니다.")
        return
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_show = img_rgb.copy()

    # 2. 얼굴 검출기 초기화
    detector = dlib.get_frontal_face_detector()

    # 3. 얼굴 검출
    faces = detector(img_rgb, 1)
    if len(faces) == 0:
        print("⚠️ 얼굴을 찾지 못했습니다.")
        return

    # 4. 각 얼굴에 사각형 그리기
    for d in faces:
        l, t, r, b = d.left(), d.top(), d.right(), d.bottom()
        cv2.rectangle(img_show, (l, t), (r, b), (0, 255, 0), 2, lineType=cv2.LINE_AA)

    # 5. 결과 시각화
    plt.imshow(img_show)
    plt.axis('off')
    plt.show()
```

```python
show_face_rectangles("images/nj.jpg")
```

```python
import cv2
import dlib
import matplotlib.pyplot as plt

def draw_face_landmarks(image_path):
    # 1. 이미지 로드
    img_bgr = cv2.imread(image_path)
    if img_bgr is None:
        print("❌ 이미지를 불러올 수 없습니다.")
        return
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_show = img_rgb.copy()

    # 2. dlib 얼굴 탐지기 및 랜드마크 모델
    predictor_path = 'shape_predictor_68_face_landmarks.dat'
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(predictor_path)

    # 3. 얼굴 검출
    faces = detector(img_rgb, 1)
    if len(faces) == 0:
        print("⚠️ 얼굴을 찾을 수 없습니다.")
        return

    # 4. 각 얼굴에 대해 랜드마크 예측
    for face in faces:
        landmarks = predictor(img_rgb, face)
        for i in range(68):  # 총 68개의 랜드마크
            x, y = landmarks.part(i).x, landmarks.part(i).y
            cv2.circle(img_show, (x, y), 2, (0, 255, 255), -1)

    # 5. 결과 출력
    plt.imshow(img_show)
    plt.axis('off')
    plt.title("Face Landmarks")
    plt.show()
```

```python
draw_face_landmarks(
    image_path="images/nj.jpg"
)
```

```python
# 각 눈에 위치한 랜드마크의 평균 좌표를 계산하여 얼굴의 좌우 각도 계산
# 각도에 맞게 수염을 회전함
```

```python
from PIL import Image, ImageOps
import math

def add_cat_whiskers_dlib(human_img_path, whisker_img_path, output_path='output.png'):
    # 이미지 불러오기
    img_bgr = cv2.imread(human_img_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(img_rgb).convert("RGBA")

    # 얼굴 검출기 및 랜드마크 예측기 초기화
    predictor_path = 'shape_predictor_68_face_landmarks.dat'
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(predictor_path)

    # 얼굴 검출
    dets = detector(img_rgb, 1)
    if len(dets) == 0:
        print("❌ 얼굴을 찾을 수 없습니다.")
        return

    # 수염 이미지 불러오기
    whisker_img_orig = Image.open(whisker_img_path).convert("RGBA")

    for d in dets:
        # 랜드마크 추출
        landmarks = predictor(img_rgb, d)
        points = [(p.x, p.y) for p in landmarks.parts()]

        # 기울기 계산
        avg_x1, avg_y1, avg_x2, avg_y2 = 0, 0, 0, 0
        
        for i in range(36,42):
            x1, y1 = points[i]
            avg_x1 += x1
            avg_y1 += y1
        avg_x1, avg_y1 = avg_x1 / 6 , avg_y1 / 6

        for i in range(42,48):
            x2, y2 = points[i]
            avg_x2 += x2
            avg_y2 += y2
        avg_x2, avg_y2 = avg_x2 / 6 , avg_y2 / 6

        dx = avg_x2 - avg_x1
        dy = avg_y2 - avg_y1
        angle = - math.degrees(math.atan2(dy, dx))  # 라디안 → 도(degree)
        print(angle)

        # 코 끝 좌표 = landmark 30번
        nose_x, nose_y = points[30]

        # 수염 이미지 크기 조정 (얼굴 폭 비례)
        face_width = d.right() - d.left()
        new_width = int(face_width * 0.6)
        new_height = int(new_width * whisker_img_orig.height / whisker_img_orig.width)
        whisker_resized = whisker_img_orig.resize((new_width, new_height))

        # 수염 회전 (중심 기준 회전)
        whisker_rotated = whisker_resized.rotate(angle, expand=True)

        # 수염 붙일 위치 계산 (회전된 크기 기준)
        wx = nose_x - whisker_rotated.width // 2
        wy = nose_y - whisker_rotated.height * 3 // 7

        # 수염 붙이기
        img_pil.paste(whisker_rotated, (wx, wy), whisker_rotated)

    # 저장 및 출력
    img_pil.save(output_path)
    print(f"✅ 결과 이미지 저장됨: {output_path}")
    plt.imshow(img_pil)
    plt.axis('off')
    plt.show()
```

```python
add_cat_whiskers_dlib(
    human_img_path="images/nj.jpg",
    whisker_img_path="images/cat-whiskers.png"
)
```
